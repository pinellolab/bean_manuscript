```python
import numpy as np
from scipy.stats import spearmanr
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import bean as be
```

```python
plt.style.use('default')
plt.rcParams["pdf.use14corefonts"] = True
matplotlib.rcParams['pdf.fonttype'] = 42 
```

```python
ldlvar = be.read_h5ad("../../results/filtered_annotated/LDLvar/bean_count_LDLvar_annotated.h5ad")
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(np.corrcoef(ldlvar.X.T, ), cbar_kws={"label":"Pearson R"})
ax.set_box_aspect(1)
ax.set_xticklabels(ldlvar.samples.index)
ax.set_yticklabels(ldlvar.samples.index, rotation=0)
fig.savefig("ldlvar_corr.pdf")
```

```python
spearmanr(ldlvar.X)[0].shape
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(spearmanr(ldlvar.X)[0], cbar_kws={"label":"Spearman R"})
ax.set_box_aspect(1)
ax.set_xticklabels(ldlvar.samples.index)
ax.set_yticklabels(ldlvar.samples.index, rotation=0)
fig.savefig("ldlvar_spearman_corr.pdf")
```
